```python
import geopandas as gpd
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
from IPython.display import HTML
import webbrowser
```

```python
df = pd.read_csv('Irridiance_Colorado.csv')
df_GHI = df.AVG_ghi
df_DHI = df.AVG_dhi
df_DNI = df.AVG_dni

def find_max_value(list):
    max = list.max()
    return max

## This is the GHI map
GHI_max = find_max_value(df_GHI)
data = list(zip(df.lata.values, df.long.values, df.AVG_ghi.values))
```

```python
coord_CO = (39.202158, -105.389073)
hmap = folium.Map(coord_CO, zoom_start=6)

hm_wide = HeatMap( data,
                   min_opacity=0.2,
                   radius=10, blur=10,
                   max_zoom=1)

hmap.add_child(hm_wide)
hmap
```
